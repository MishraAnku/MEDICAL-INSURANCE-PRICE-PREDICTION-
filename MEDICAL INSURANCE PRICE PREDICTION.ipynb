    # Assuming a simple parabolic equation: y = ax^2 + bx + c
    a = 0.01
    b = 5
    c = 1000
    parabola_x = range(int(min(x)), int(max(x))+1)
    parabola_y = [a*x_val**2 + b*x_val + c for x_val in parabola_x]

    # Set labels and title
    plt.xlabel('Age')
    plt.ylabel('expenses')
